```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([10,20,30,40,50,60])
y = np.array([1,10,3,40,2,4])
plt.plot(x,y)
plt.show()
```

```python
# legend()
# labels()
# Title()
```

```python
plt.plot(x,y,marker='o')
plt.show()
```

```python
# Format strings (fmt) [marker:line:color]
# line type are of four types [- , -. , : , --]
plt.plot(x,y,"o-.g") # Here marker type is 'o', color is green and by default line type is '-.'
plt.show()
```

```python
plt.plot(x,y,marker='<',ms=10) # ms(marker size) is increased here
plt.show()
```

```python
# Formatting lines
plt.plot(x,y,lw='10.5',c='#92FBA9',marker='o',ms=20,ls='--')
plt.show() # Here lw is line width, c is color , ms = marker size , ls= line style.
```

```python
%matplotlib inline
plt.plot(x,y,"o--g",ms=20,lw='10.5')
# This is a short form of above presented graph.
```

```python
# To plot multiple graphs.
y1 = np.array([10,20,30,40,50,60])
y2 = np.array([1,10,3,40,2,4])
plt.plot(y1)
plt.plot(y2)
plt.show()
```

```python
y1 = np.array([10,20,30,40,50,60])
y2 = np.array([1,10,3,40,2,4])
plt.plot(y1,marker= 'o')
plt.plot(y2,ls='-.')
plt.show()
```

```python
x1 = np.array([7,8,9,10,11,12])
y1 = np.array([10,20,30,40,50,60])
x2 = np.array([1,2,3,4,5,6])
y2 = np.array([1,10,3,40,2,4])
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

```python
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.xlabel("Year")
plt.ylabel("Sales")
plt.legend(["Maruti","Tata"])
plt.title("Sales in Year")
plt.show()
```

```python
# To customize labels and title.
plt.plot(x1,y1)
font_x = {"family":"serif","color":"#F51010","size":15}
font_y = {"family":"serif","color":"#FA6010","size":20}
plt.xlabel("Year",fontdict=font_x)
plt.ylabel("Sales",fontdict=font_y)
plt.legend(["Maruti","Tata"])
plt.title("Sales in Year")
plt.show()
```

```python
plt.plot(x1,y1)
font_x = {"family":"serif","color":"#F51010","size":15}
font_y = {"family":"serif","color":"#FA6010","size":20}
plt.xlabel("Year",fontdict=font_x)
plt.ylabel("Sales",fontdict=font_y)
plt.legend(["Maruti","Tata"])
plt.title("Sales in Year", loc='left') # loc to change position of title.
plt.show()
```

# TO DRAW GRIDS

```python
plt.plot(x1,y1)
font_x = {"family":"serif","color":"#F51010","size":15}
font_y = {"family":"serif","color":"#FA6010","size":20}
plt.xlabel("Year",fontdict=font_x)
plt.ylabel("Sales",fontdict=font_y)
plt.grid() # We can pass parameters also in grid function like grid(axis='x')
plt.legend(["Maruti","Tata"])
plt.title("Sales in Year",loc="Right")
plt.show()
```

```python
plt.plot(x1,y1)
font_x = {"family":"serif","color":"#F51010","size":15}
font_y = {"family":"serif","color":"#FA6010","size":20}
plt.xlabel("Year",fontdict=font_x)
plt.ylabel("Sales",fontdict=font_y)
plt.grid(color='black',lw=1,ls='-.',axis='y') # This is how we can customize grid lines.
plt.legend(["Maruti","Tata"])
plt.title("Sales in Year",loc="Right")
plt.show()
```

#  # SUBPLOT

```python
x1 = np.array([7,8,9,10,11,12])
y1 = np.array([10,20,30,40,50,60])
x2 = np.array([1,2,3,4,5,6])
y2 = np.array([1,10,3,40,2,4])
# Subplot(rows,columns,index)
# Do subplot before plot.
plt.subplot(2,2,1)
plt.plot(x1,y1,lw=1,ls='--',marker='+',ms=12)

plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.show()
```

```python
plt.legend(["Sales"])
plt.subplot(1,2,1)
plt.plot(x1,y1,lw=1,ls='--',marker='+',ms=12)
plt.xlabel("Sales")

#plt.legend(["Sales2"])
plt.subplot(1,2,2)
plt.plot(x2,y2)
plt.xlabel("Year")
plt.show()

```

```python
a = np.array([5,7,8,7,2,17,2,6,9])
b = np.array([99,86,87,88,111,115,59,69,75])
size = np.array([20,50,100,200,500,165,99,80,400])
color=np.array(["red","green","yellow","blue","violet","black","red","pink","green"])
plt.scatter(a,b)
plt.show()
```

```python
plt.scatter(a,b,color=color,s=size)
plt.title("Random",loc="left")
plt.xlabel("colorur")
plt.ylabel("sizee")
plt.show()
```

```python
# Colormap :- Ranges betwwen 1 to 100
plt.scatter(a,b,c = np.arange(9),cmap="plasma")
plt.colorbar()
plt.show()
```

```python
plt.scatter(a,b,c=np.arange(9),cmap="viridis",alpha=0.5) # Here alpha is transparency and its value ranges between 0 and 1.
plt.colorbar()
plt.show()
```

# BAR GRAPH

```python
d = np.array([2017,2018,2019,2020,2021,2022])
e = np.array([100,200,300,400,200,500])
plt.bar(d,e)
plt.show()
```

```python
f = np.array(['2017','2018','2019','2020','2021','2022'])
g = np.array([100,200,300,400,200,500])
plt.barh(f,g) # barh for printing horizontal bar graph.
plt.show()
```

```python
f = np.array(['2017','2018','2019','2020','2021','2022'])
g = np.array([100,200,300,400,200,500])
plt.bar(f,g,color='red',width=0.5)
plt.show()
```

# Pie Chart

```python
x1 = np.array([25,25,15,35])
plt.pie(x1)
plt.show()
```

```python
# To define labels:-
la = ["Delhi","Bhiwani","Rohtak","Dehradun"]
plt.pie(x1,labels=la)
plt.show()
```

```python
# To make it more attractive. Explode it.
exolode = [0.2,0.1,0,0]
plt.pie(x1,labels=la,explode=exolode)
plt.show()
```

```python
color=np.array(["red","green","yellow","blue"])
plt.pie(x1,labels=la,explode=exolode,shadow = True,colors=color)
plt.show()
```
