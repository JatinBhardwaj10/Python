```python
df = sb.load_dataset("tips")
df.sample(10).head()
```

```python
# Realtional plot (relplot).
sb.relplot(x="total_bill",y="tip",data=df)
plt.show()
```

```python
sb.relplot(x="total_bill",y="tip",data=df,hue='smoker')
plt.show()
```

```python
sb.relplot(x="total_bill",y="tip",data=df,hue='smoker',style='sex')
plt.show()
```

```python
sb.relplot(x="total_bill",y="tip",data=df,hue='smoker',style='sex',palette='viridis')
plt.show()
```

```python
sb.relplot(x="total_bill",y="tip",data=df,row='smoker',col='sex',palette='viridis')
plt.show()
```

```python
sb.catplot(x="day",y='total_bill',data=df)
plt.show()
```

```python
sb.catplot(x="day",y='total_bill',data=df,jitter=False)
plt.show()
```

```python
sb.catplot(x="day",y='total_bill',data=df,kind='swarm')
plt.show()
```

```python
sb.boxplot(x="day",y='total_bill',data=df)
plt.show()
# We can also do :- sb.catplot(x="day",y='total_bill',data=df,kind='box')
```

```python
# In boxen plot we pnly focus on 25%ile,50%ile,75%ile.
sb.catplot(x="day",y='total_bill',data=df,kind='boxen')
plt.show()
```

```python
sb.catplot(x="day",y='total_bill',data=df,kind='violin')
plt.show()
```

```python
sb.catplot(x="day",y='total_bill',data=df,kind='violin',hue='smoker',split=True)
plt.show()
```

```python
sb.catplot(x="day",data=df,kind='count',hue='smoker',col='sex')
plt.show()
```

```python
sb.catplot(x="day",y='total_bill',data=df,kind='bar',hue='smoker')
plt.show()
```

```python
x = np.random.normal(size=200)
sb.distplot(x)
# This line on the graph is knows as kde (kernel density estimation).
```

```python
x = np.random.normal(size=200)
sb.distplot(x,hist=False,rug=True,kde=False)
```

#    Regression Plot

```python
sb.regplot(x="total_bill",y="tip",data=df)
plt.show()
# This line is known as regression line and shaded region is standard deviation.
```

```python
c1 = sb.load_dataset('anscombe')
c1.head()
```

```python
sb.lmplot(x='x',y='y',data=c1)
plt.show()
```

```python
sb.lmplot(x='x',y='y',data=c1.query("dataset=='I'"))
plt.show()
```

```python
sb.lmplot(x='x',y='y',data=c1.query("dataset=='I'"),order=3)
plt.show()
```

```python
sb.lmplot(x='x',y='y',data=c1.query("dataset=='II'"),order=2)
```

```python
# To change labels and title
a = sb.lmplot(x='x',y='y',data=c1.query("dataset=='II'"),order=2)
a.set(xlabel='X-values',
      ylabel='Y-values',
title="X and Y"    
)
plt.show()
```

# Matrix Plots

```python
df = sb.load_dataset('flights')
df.head()
```

```python
p = df.pivot('month','year','passengers')
p.head()
```

```python
sb.heatmap(p)
plt.show()
```

```python
sb.heatmap(p,annot=True,fmt='d')
plt.show()
```

```python
sb.heatmap(p,annot=True,fmt='d',lw=1) # lw is line_width.
plt.show()
```

# Cluster Map

```python
df = sb.load_dataset('flights')
df1 = df.drop(columns='month')
```

```python
sb.clustermap(df1)
plt.show()
```


# Grids

```python
tip = sb.load_dataset("tips")
tip.head()
```

```python
tp = sb.FacetGrid(data=tip,row='smoker',col='time')
tp.map(sb.distplot,'total_bill')
```

```python
tp = sb.FacetGrid(data=tip,row='smoker',col='time')
tp.map(sb.scatterplot,'total_bill','tip',color='green')
```
